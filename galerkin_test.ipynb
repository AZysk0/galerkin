```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import scipy.optimize as optimize

```

```python
pi = np.pi
m = 3
```

```python
def w(j, x):
    """Orthonormal eigenfunction on (0, pi)."""
    return np.sqrt(2 / pi) * np.sin(j * x)

lambdas = np.array([j**2 for j in range(1, m + 1)])

print("lambdas =", lambdas)
```

```python
def inner(f, g):
    """L2 inner product on (0, pi)."""
    value, error = integrate.quad(lambda x: f(x) * g(x), 0, pi)
    return value

def u0(x):
    return 1.0 * w(1, x) + 0.5 * w(2, x)

def b(x):
    return 0.3 * w(1, x) - 0.2 * w(3, x)

def project_coeffs(f):
    """Return coefficients ((f,w1),...,(f,wm))."""
    return np.array([
        inner(f, lambda x, j=j: w(j, x))
        for j in range(1, m + 1)
    ])

c0 = project_coeffs(u0)
beta = project_coeffs(b)

print("c0   =", c0)
print("beta =", beta)
```
